# Inserting data into sql data base using open command

```python
import mysql.connector as connection
import pandas as pd
import csv
```

```python
mydb = connection.connect(host="localhost", database = 'shubham1',user="root", passwd="Biology123@",use_pure=True)
cur = mydb.cursor()
cur.execute("Create Table Glass (RI float(10,5),Na float(10,5),Mg float(10,5),Al float(10,5),Si float(10,5),K float(10,5),Ca float(10,5),Ba float(10,5),Fe float(10,5),Type int(2))")
```

```python
# using this method we can read data from any txt file
with open("glass.csv",'r') as data:
    for i in data: 
        print(i)
```

```python
# We are updating our new table name Glass
with open("glass.csv",'r') as data:
    next(data) # skipping first data
    data_csv = csv.reader(data, delimiter = '\n') # this will create new data when ever new line detected
    print(data_csv) # this will just to verify that wether it create any data vairable or not
    for i in enumerate(data_csv): 
        print(i)
        for j in i[1]:
            cur.execute("Insert into Glass values ({data})".format(data = (j)))
    print("Task Completed")
mydb.commit()
```

```python
# Taking value which have type == 6
cur.execute("Select * From Glass where Type = 6")
for i in cur.fetchall():
    print(i)
```
