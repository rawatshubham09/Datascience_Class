## Pareto Distribution

```python
import numpy as np
from scipy import stats

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generating data
pare_data = np.random.exponential(size = 1000)
pare_data
```

```python
# Appling BoxCox transformation and generating new data with its lambda value:
format_data , lambda_val = stats.boxcox(pare_data)
```

```python
# Spliting ploting finction into two:
fig ,ax = plt.subplots(1,2)

#ploting orignal data and nornaldata side by side:
#Pareto data
sns.distplot(pare_data, hist = False, kde =True, kde_kws = {'shade' : True, 'linewidth' : 2},
             label = "Non-Normal", color = "green", ax = ax[0]);
#After box_cox distribution
sns.distplot(format_data, hist = False, kde =True, kde_kws = {'shade' : True, 'linewidth' : 2},
             label = "Normal", color = "red", ax = ax[1]);

#rescaling subplot size:
fig.set_figheight(5)
fig.set_figwidth(10)

print("Lambda value use for transform :", lambda_val)
```

# Defineing Function Related this topic:

```python
# define a log_transformation for given col in any data frame:

def log_transformation(df, colns):
    for col in colns:
        df[col] = np.log(df[col]+1)
    return df
```

```python
# define a inverse log transformation:
def inverse_log_transformation(df, colns):
    for col in colns:
        df[col] = np.exp(df[col])-1
    return df
```
