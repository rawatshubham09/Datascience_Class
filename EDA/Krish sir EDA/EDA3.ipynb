```python
# creating function to convert duration into minutes
def to_min(st):
    data = st.split(" ")
    minut = 0
    for i in data:
        if i[-1]=='h':
            minut += int(i[:-1])*60
        if i[-1]=='m':
            minut += int(i[:-1])
    return minut
```

```python
df.Duration = df.Duration.apply(to_min)
```

```python
# Handling Airline Source Destination and Additional info
df.Airline.value_counts()
```

```python
def to_airline(st):
    name = ['Multiple carriers Premium economy','Vistara Premium economy','Jet Airways Business','Trujet']
    if st in name:
        return 'Others'
    return st
df.Airline = df.Airline.apply(to_airline)
df.Airline.value_counts()
```

```python
def to_add_info(st):
    name = ['1 Long layover', 'Change airports', 'Business class', 'No Info','1 Short layover', 'Red-eye flight', '2 Long layover']
    return 'Other' if st in name else st
```

```python
df.Additional_Info=df.Additional_Info.apply(to_add_info)
df.Additional_Info.value_counts()
```
