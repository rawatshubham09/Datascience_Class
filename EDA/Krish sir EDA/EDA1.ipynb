```python
final_df.Country.value_counts()
```

```python
final_df.Country.value_counts().index
```

```python
country_name = final_df.Country.value_counts().index
country_val = final_df.Country.value_counts().values
```

```python
# Pi chart top 4 country
plt.figure(figsize=(10,10))
plt.pie(country_val[:4],labels=country_name[:4], autopct='%1.2f%%');
```

Observation: Zomato Max Transitaion are from India then followed by USA and UK

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index()
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings.columns
```

- Observations:
    1. 4.5-4.9 ---> Excellent
    2. 4.0-4.4 ---> VeryGood
    3. 3.5-3.9 ---> Good
    4. 2.5-3.4 ---> Average
    1. 1.8-2.4 ---> Poor
    1. 0 ---> Not Rated

```python
sns.barplot(x='Aggregate rating',y='Rating Count', data = ratings);
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count', data = ratings, hue='Rating color', palette=['black','red', 'orange', 'yellow','green','green']);
```

- Observations:
    1. Max People dont like to Rate app
    2. Avg people Rate between 2.5 and 3.9
