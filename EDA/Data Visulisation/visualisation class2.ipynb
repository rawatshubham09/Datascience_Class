```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# how many people servived as compart to there sex:
titanic.iplot(x = 'sex', y = 'survived', kind = 'bar')
```

```python
#relation between age and fare:
titanic.iplot(x = 'fare', y = 'age', mode = 'markers')
```

There is no relation between age and fare:

```python
# adding new feature family = sibsp + parch

titanic['family'] = titanic['sibsp']+ titanic['parch']
titanic = titanic.drop(['sibsp', 'parch'], axis = 1)
titanic.head()
```

```python
# relation between family size and fare:
titanic.iplot(x='pclass', y = 'fare', kind = 'scatter3d', z = 'family')
```
